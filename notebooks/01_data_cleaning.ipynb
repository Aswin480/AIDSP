```python
import pandas as pd
import numpy as np

```

```python
enrol = pd.read_csv("../data/raw/aadhaar_enrolment.csv")
demo  = pd.read_csv("../data/raw/aadhaar_demographic_updates.csv")
bio   = pd.read_csv("../data/raw/aadhaar_biometric_updates.csv")

print(enrol.shape)
print(demo.shape)
print(bio.shape)
```

```python
print(enrol.columns)
print(demo.columns)
print(bio.columns)
```

```python
enrol['date'] = pd.to_datetime(enrol['date'], dayfirst=True, errors='coerce')
demo['date']  = pd.to_datetime(demo['date'], dayfirst=True, errors='coerce')
bio['date']   = pd.to_datetime(bio['date'], dayfirst=True, errors='coerce')
```

```python
print(enrol['date'].head())
print(demo['date'].head())
print(bio['date'].head())
```

```python
enrol = enrol.sort_values(['state', 'date'])
demo  = demo.sort_values(['state', 'date'])
bio   = bio.sort_values(['state', 'date'])
```

```python
print(enrol.isnull().sum())
print(demo.isnull().sum())
print(bio.isnull().sum())
```

```python
enrol.to_csv("../data/processed/enrolment_clean.csv", index=False)
demo.to_csv("../data/processed/demographic_clean.csv", index=False)
bio.to_csv("../data/processed/biometric_clean.csv", index=False)
```
