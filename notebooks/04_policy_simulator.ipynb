```python
import pandas as pd
import numpy as np
```

```python
forecast = pd.read_csv("../results/state_risk_forecast.csv")

forecast.head()
```

```python
policy_scenarios = {
    "low_intervention": 0.90,     # 10% risk reduction
    "medium_intervention": 0.70,  # 30% risk reduction
    "high_intervention": 0.50     # 50% risk reduction
}
```

```python
for policy, factor in policy_scenarios.items():
    forecast[policy] = forecast['predicted_risk_score'] * factor

forecast.head()
```

```python
forecast.to_csv(
    "../results/policy_simulation_output.csv",
    index=False
)

print("✅ POLICY SIMULATION SAVED")
```
